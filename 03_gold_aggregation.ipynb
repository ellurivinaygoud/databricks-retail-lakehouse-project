```python
## Read data from silver table
df = spark.table("silver_Ecommerce_Data")

display(df)
```

```python
## Calculating Gold Table

from pyspark.sql.functions import sum,count

gold_df = df.groupBy("Country") \
            .agg(count("InvoiceNo").alias("TotalOrders"),
                 sum("Quantity").alias("TotalQuantity"),
                 sum("UnitPrice").alias("Revenue"))

gold_df.write.format("delta").mode("overwrite").saveAsTable("gold_Ecommerce_Data")
```
