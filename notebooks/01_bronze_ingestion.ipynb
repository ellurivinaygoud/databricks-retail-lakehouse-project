```python
## Reading the data from the source table
df = spark.read.table("workspace.default.ecommerce_data")
display(df.describe())
```

```python
## Writing the data to the target table
df.write.format("delta").mode("overwrite").saveAsTable("bronze_Ecommerce_Data")
```

```python
## Identifying null columns in the dataframe
from pyspark.sql.functions import col, sum

null_counts = df.select(
    [sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]
)

null_counts.show()
```

```python
## Verifying null columns count in the dataframe
null_summary = null_counts.collect()[0].asDict()

columns_with_nulls = [k for k, v in null_summary.items() if v > 0]

columns_with_nulls
```
