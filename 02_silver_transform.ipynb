```python
## Removing the nulls from the bronze table 

from pyspark.sql.functions import col, date_format, to_timestamp

df = spark.table("bronze_Ecommerce_Data")
clean_df = df.withColumn(
    "InvoiceDate",
    date_format(to_timestamp("InvoiceDate", "M/d/yyyy H:mm"), "M/d/yyyy")
).filter(col("CustomerID").isNotNull())
clean_df.write.format("delta").mode("overwrite").saveAsTable("silver_Ecommerce_Data")
```
